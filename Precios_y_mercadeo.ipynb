# Clase de clasificaci√≥n de riesgo

A partir del conocimiento del tipo de riesgo que tiene un cliente podemos determinar un precio a ese riesgo mediante polizas o seguros.

```python
import numpy as np # Esto importa el paquete
import matplotlib.pyplot as plt

# Definir el rango de x
x = np.linspace(1, 50, 1000)

# Funciones
def funcion_amante(x):
    return np.exp(x)

def funcion_averso(x):
    return np.sqrt(x)

def funcion_neutral(x):
    return 0.5 * x + 1

def funcion_logaritmica(x):
    return np.log(x)

# Crear la figura y los subplots
fig, axs = plt.subplots(1, 4, figsize=(15, 5))

# Gr√°fico 1: Funci√≥n Amante
axs[0].plot(x, funcion_amante(x), label="Amante")
axs[0].set_title("Funci√≥n Amante al riesgo")
axs[0].legend()

# Gr√°fico 2: Funci√≥n Averso
axs[1].plot(x, funcion_averso(x), label="Averso")
axs[1].set_title("Funci√≥n Averso al riesgo")
axs[1].legend()

# Gr√°fico 3: Funci√≥n Neutral
axs[2].plot(x, funcion_neutral(x), label="Neutral al Riesgo")
axs[2].set_title("Funci√≥n Neutral al Riesgo")
axs[2].legend()

# Gr√°fico 4: Funci√≥n Logaritmica
axs[3].plot(x, funcion_logaritmica(x), label="Con logaritmo Riesgo")
axs[3].set_title("Funci√≥n logaritmo")
axs[3].legend()


# Ajustar el espaciado entre subplots
plt.tight_layout()

# Mostrar los gr√°ficos
plt.show()
```

```python
# Datos de las tres loter√≠as
loteria1 = [(100000, 0.5), (-50000, 0.5)]
loteria2 = [(200000, 0.5), (-100000, 0.5)]
loteria3 = [(20000000, 0.5), (-10000000, 0.5)]

# Funci√≥n para calcular el valor esperado
def cal_valor_esperado(loteria):
    valor_esperado = 0
    for resultado, probabilidad in loteria:
        valor_esperado += resultado * probabilidad
    return valor_esperado

# Calcular valor esperado para cada loter√≠a
valor_esperado1 = cal_valor_esperado(loteria1)
valor_esperado2 = cal_valor_esperado(loteria2)
valor_esperado3 = cal_valor_esperado(loteria3)

# Imprimir los resultados
print("Valor Esperado Loter√≠a 1:", valor_esperado1)
print("Valor Esperado Loter√≠a 2:", valor_esperado2)
print("Valor Esperado Loter√≠a 3:", valor_esperado3)

# Encontrar la loter√≠a con el valor esperado m√°s alto
loterias = {valor_esperado1: "Loter√≠a 1", valor_esperado2: "Loter√≠a 2", valor_esperado3: "Loter√≠a 3"}
mejor_loteria = loterias[max(valor_esperado1, valor_esperado2, valor_esperado3)]

print("La mejor loter√≠a es:", mejor_loteria)
```

```python
# Datos de las tres loter√≠as de Mateo
loteria1 = [(10100000, 0.5), (9950000, 0.5)]
loteria2 = [(10200000, 0.5), (9900000, 0.5)]
loteria3 = [(30000000, 0.5), (0, 0.5)]

# Funci√≥n para calcular la utilidad esperada
def cal_utilidad_esperada(loteria):
    utilidad_esperada = 0
    for resultado, probabilidad in loteria:
        utilidad_esperada += (resultado)**0.5 * probabilidad
    return utilidad_esperada

# Calcular utilidad esperada para cada loter√≠a
utilidad_esperada1 = cal_utilidad_esperada(loteria1)
utilidad_esperada2 = cal_utilidad_esperada(loteria2)
utilidad_esperada3 = cal_utilidad_esperada(loteria3)

# Imprimir los resultados de utilidad esperada
print("Utilidad Esperada Loter√≠a 1:", utilidad_esperada1)
print("Utilidad Esperada Loter√≠a 2:", utilidad_esperada2)
print("Utilidad Esperada Loter√≠a 3:", utilidad_esperada3)

# Encontrar la loter√≠a con la utilidad esperada m√°s alta
utilidad_loterias = {utilidad_esperada1: "Loter√≠a 1", utilidad_esperada2: "Loter√≠a 2", utilidad_esperada3: "Loter√≠a 3"}
Mejor_loteria = utilidad_loterias[max(utilidad_esperada1, utilidad_esperada2, utilidad_esperada3)]

print("La mejor loter√≠a en t√©rminos de utilidad esperada es la:", Mejor_loteria)
```

```python
# Datos de las tres loter√≠as
loteria1 = [(810000, 0.1), (360000, 0.5), (110000, 0.4)]
loteria2 = [(250000, 1)] # Modificar
loteria3 = [(450000, 0.3), (170000, 0.7)]

# Funci√≥n de utilidad
def funcion_utilidad(x):
    return np.log(x)

# Funci√≥n para calcular el valor esperado
def cal_valor_esperado(loteria):
    valor_esperado = 0
    for resultado, probabilidad in loteria:
        valor_esperado += resultado * probabilidad
    return valor_esperado

# Calcular valor esperado para cada loter√≠a
valor_esperado1 = cal_valor_esperado(loteria1)
valor_esperado2 = cal_valor_esperado(loteria2)
valor_esperado3 = cal_valor_esperado(loteria3)

# Funci√≥n para calcular la utilidad esperada
def cal_utilidad_esperada(loteria):
    utilidad_esperada = 0
    for resultado, probabilidad in loteria:
        utilidad_esperada += np.log(resultado) * probabilidad
    return utilidad_esperada

# Calcular utilidad esperada para cada loter√≠a
utilidad_esperada1 = cal_utilidad_esperada(loteria1)
utilidad_esperada2 = cal_utilidad_esperada(loteria2)
utilidad_esperada3 = cal_utilidad_esperada(loteria3)

# Funci√≥n de utilidad
def funcion_utilidad(x):
    return np.exp(x) # al ser logaritmo se aplica el exponencial (e)

# Calcular equivalente de certeza como la utilidad de un monto seguro equivalente
equivalente_certeza1 = funcion_utilidad(utilidad_esperada1)
equivalente_certeza2 = funcion_utilidad(utilidad_esperada1)
equivalente_certeza3 = funcion_utilidad(utilidad_esperada1)

# Calcular la prima de riesgo como la diferencia entre el valor esperado y el equivalente de certeza
Prima_riesgo1 = valor_esperado1 - equivalente_certeza1
Prima_riesgo2 = valor_esperado2 - equivalente_certeza2
Prima_riesgo3 = valor_esperado3 - equivalente_certeza3

# Imprimir los resultados de la prima de riesgo
print("Prima de Riesgo Loter√≠a 1:", Prima_riesgo1)
print("Prima de Riesgo Loter√≠a 2:", Prima_riesgo2)
print("Prima de Riesgo Loter√≠a 3:", Prima_riesgo3)
```
